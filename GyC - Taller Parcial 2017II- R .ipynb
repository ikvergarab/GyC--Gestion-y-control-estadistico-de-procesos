**2.** Luego del análisis realizado en el punto anterior, el encargado le comenta que tienen un cliente muy grande en el exterior que le compra las Iris, pero con ciertas características, para cada una de las siguientes condiciones indique por medio de la inferencia estadística si se cumplen o no. 

**a.** El largo del sépalo (Sepal.Length) de todas las Iris, independiente de su especie, debe ser superior a 2.2 mm. 

```r
# Se debe indicar primero cuales son nuestras pruebas. Se trabajara un alpha de 0.05
# H0: mu=2.2
# H1: mu>2.2
test1=t.test(Sepal.Length,  #conjunto de datos al que se le realizará la prueba
             mu=2.2, # valor a probar
             conf.level = 0.95, # coeficiente
             alternative = "greater") # probar desigualdad del tipo mayor que
test1
```

#### Como alpha es mayor que el vp rechazo H0, por tanto con una confianza del 95% la media largo del sépalo de todas las Iris es estadisticamente mayor a 2.2 mm


**b.** El ancho de los pétalos (Petal. Width) de todas las Iris, independiente de su especie, debe ser superior a 1 mm.

```r
# H0: mu=1
# H1: mu>1
test2=t.test(Petal.Width, 
             mu=1,
             conf.level = 0.95,
             alternative = "greater")
test2
```

#### Como alpha es mayor que el vp rechazo H0, por tanto con una confianza del 95% la media ancho de lo petalos de todas las Iris es estadisticamente mayor a 1 mm


**c.** Para la especie setosa, el cliente requiere que las medidas del largo de los pétalos no excedan los 2 mm. 

```r
test3=t.test(Petal.Length[Species=="setosa"], # siempre verifiquen como esta escrito el nombre en 
                                              # la columna, deben de escribirlo exactamente igual
             mu=2,
             conf.level = 0.95,
             alternative = "less")
test3
```

####  Dado que el alpha es mayor qur el vp rechazo la hipotesis nula, por tanto con una confianza del 95% para la especie setosa el largo de los petalos no exceden estadisticamente los 2 mm

**d.** Para la especie versicolor, se necesita que la medida del ancho del pétalo supere 1mm.
